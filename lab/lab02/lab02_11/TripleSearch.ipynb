# Задание 14: Троичный поиск

Трои́чный по́иск (Тернарный поиск) — это метод в информатике для поиска максимумов и минимумов функции, которая либо сначала строго возрастает, затем строго убывает, либо наоборот. Троичный поиск определяет, что минимум или максимум не может лежать либо в первой, либо в последней трети области, и затем повторяет поиск на оставшихся двух третях. Троичный поиск демонстрирует парадигму программирования «разделяй и властвуй».

# Достоинства и недостатки алгоритма

Алгоритм троичного поиска достоинства:

Более эффективный, чем линейный поиск, но менее эффективный, чем бинарный поиск.
Работает на отсортированных массивах.
Более универсальный, чем бинарный поиск, так как может использоваться для поиска в неполностью отсортированных массивах.
В отличие от бинарного поиска, троичный поиск может использоваться для поиска нескольких элементов с одинаковым значением.
Алгоритм троичного поиска недостатки:

Большая вычислительная сложность и длинное время выполнения при поиске в массивах большой длины.
Не подходит для поиска в массивах, в которых значения повторяются многократно.
Не всегда удобен для понимания и реализации.
В случае, если элемент находится в крайней левой или крайней правой части массива, алгоритм троичного поиска может потребовать большого количества итераций для его обнаружения.

В целом, троичный поиск является хорошим выбором для поиска значений в небольших отсортированных массивах или в случаях, когда важно найти все элементы с одинаковым значением. Однако, в случае поиска в больших массивах, бинарный поиск может оказаться более эффективным.

# Реализация алгоритма

```python
def ternary_search(l, r, eps, f):
    l1 = l
    r1 = r
    while r - l > eps:
        m1 = l + (r - l) / 3
        m2 = r - (r - l) / 3
        if f(m1) < f(m2):
            r = m2
        else:
            l = m1
    min = (l + r) / 2
    while r1 - l1 > eps:
        m1 = l1 + (r1 - l1) / 3
        m2 = r1 - (r1 - l1) / 3
        if f(m1) < f(m2):
            l1 = m1
        else:
            r1 = m2
    max = (l1 + r1) / 2
    return min, max
```

# Тестирование алгоритма

```python
def f(m):
    return 12*(m**3) + 12*(m**3)

result = ternary_search(-10, 10, 10**(-10), f)
print(result)
```

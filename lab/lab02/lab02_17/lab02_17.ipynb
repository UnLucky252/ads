#### 2.Подготовить теоретическое описание алгоритмов согласно номеру индивидуального варианта.

Линейный поиск - это метод сортировки, который проходит по всем элементам неупорядоченного массива данных и ищет заданный элемент.

Алгоритм линейного поиска заключается в следующих шагах:

Пройти по всем элементам массива, начиная с первого элемента.
Если текущий элемент соответствует искомому элементу, то возвратить его индекс.
Если искомый элемент не найден после прохода по всем элементам массива, то вернуть специальное значение, обозначающее отсутствие элемента (например, -1).
Преимущества линейного поиска в том, что он прост в реализации и может быть использован для поиска элементов в неупорядоченных массивах данных. Однако, он имеет низкую эффективность для больших массивов данных, так как время поиска может быть пропорционально количеству элементов в массиве.

Также стоит отметить, что линейный поиск не работает для отсортированных массивов данных, так как существуют более эффективные алгоритмы для этого случая, такие как бинарный поиск.

#### 4.Представить описание алгоритмов на псевдокоде.

```python
int linear_search(int arr[], int n, int target) {
    for(int i = 0; i < n; i++) {
        if(arr[i] == target) {
            return i; // возвращаем индекс элемента в массиве
        }
    }
    return -1; // возвращаем -1, если элемент не найден в массиве
}
```

#### 5.Описать достоинства и недостатки алгоритмов.

Достоинства:
Простота: линейный поиск является простым алгоритмом, который не требует специальных знаний и может быть легко реализован на любом языке программирования.

Универсальность: линейный поиск может быть использован для поиска элементов в любом массиве, включая несортированные массивы.

Эффективность для маленьких массивов: для небольших массивов линейный поиск может быть более эффективным, чем более сложные алгоритмы поиска.

Однако у линейного поиска есть и некоторые недостатки:

Низкая производительность для больших массивов: линейный поиск выполняется за O(n) времени, где n - это количество элементов в массиве. Это означает, что для больших массивов время выполнения может быть довольно долгим.

Неэффективность для отсортированных массивов: если массив отсортирован, более эффективным алгоритмом будет бинарный поиск, который выполняется за O(log n) времени.

Невозможность выйти досрочно: линейный поиск должен проверить каждый элемент в массиве, даже если искомый элемент находится в начале массива. Это означает, что линейный поиск не может выйти досрочно, как это делает бинарный поиск, когда он находит искомый элемент.

#### 6-7. Реализовать алгориты согласно номеру индивидуального варианта. Протестировать корректность реализации алгоритма

```python
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return None

# Тест 1 - элемент найден
arr1 = [1, 2, 3, 4, 5]
target1 = 3
assert linear_search(arr1, target1) == 2

# Тест 2 - элемент не найден
arr2 = [1, 2, 3, 4, 5]
target2 = 6
assert linear_search(arr2, target2) == None

# Тест 3 - повторяющиеся элементы
arr3 = [1, 2, 3, 4, 5, 3]
target3 = 3
assert linear_search(arr3, target3) == 2

# Тест 4 - пустой список
arr4 = []
target4 = 1
assert linear_search(arr4, target4) == None
```

#### задание 1. Алгоритм Бойера — Мура.

#### 2.Подготовить теоретическое описание алгоритмов согласно номеру индивидуального варианта.

Алгоритм Бойера-Мура является алгоритмом поиска подстроки в строке, который основан на предположении, что последний символ шаблона, который не совпадает с искомой подстрокой, обеспечивает наибольшую информацию о том, где может находиться искомая подстрока.

Основные шаги алгоритма Бойера-Мура:

Подготовка таблицы смещений:

Для каждого символа в алфавите строится таблица смещений, которая показывает наибольшее смещение, на которое можно сдвинуть шаблон, если он не совпадает с символом, находящимся на данном месте в строке.
Если символ не встречается в шаблоне, то сдвигается на длину шаблона.
Поиск подстроки в строке:

Сравниваем символы шаблона с символами строки, начиная с конца шаблона и двигаясь к началу.
Если символы совпадают, продолжаем сравнение до тех пор, пока не найдем полное совпадение.
Если символ не совпадает, используем таблицу смещений, чтобы определить наибольшее смещение, на которое можно сдвинуть шаблон, чтобы продолжить поиск.
Если полное совпадение найдено, возвращаем индекс, где начинается искомая подстрока.
Алгоритм Бойера-Мура показывает высокую эффективность на практике, особенно при поиске подстроки в длинных текстах с большим алфавитом.

#### 4.Представить описание алгоритмов на псевдокоде.

```python
void search(string text, string pattern) {
    int m = pattern.length();
    int n = text.length();

    // Шаг 1: Создаем таблицу смещений
    int badchar[256];
    memset(badchar, -1, sizeof badchar);
    for (int i = 0; i < m; i++) {
        badchar[(int)pattern[i]] = i;
    }

    // Шаг 2: Поиск подстроки
    int s = 0; // смещение шаблона относительно текста
    while (s <= n - m) {
        int j = m - 1;

        // Проверяем символы с конца шаблона до начала,
        // пока они совпадают с символами соответствующей позиции в тексте
        while (j >= 0 && pattern[j] == text[s+j]) {
            j--;
        }

        if (j < 0) {
            // Подстрока найдена
            cout << "Подстрока найдена с позиции " << s << endl;
            s += (s+m < n) ? m-badchar[(int)text[s+m]] : 1;
        } else {
            // Смещаем шаблон на максимальное возможное количество символов
            s += max(1, j - badchar[(int)text[s+j]]);
        }
    }
}
```

Этот код принимает две строки - text и pattern - и находит все вхождения pattern в text. На первом шаге (создание таблицы смещений), алгоритм заполняет массив badchar, где для каждого символа из pattern хранится его последнее вхождение в pattern. Затем алгоритм начинает поиск подстроки, начиная с начала текста и сравнивая символы с конца шаблона до начала. Если символы не совпадают, алгоритм смещает шаблон на максимальное возможное количество символов, используя таблицу смещений badchar. Если символы совпадают, алгоритм продолжает сравнение символов в обратном порядке, пока не достигнет начала шаблона или пока не найдет несовпадающий символ. Если весь шаблон совпадает с подстрокой текста, алгоритм сообщает о нахождении подстроки и смещает шаблон на следующее возможное место.

#### 5.Описать достоинства и недостатки алгоритмов.

Достоинства:

Эффективность: Алгоритм Бойера-Мура - это один из самых эффективных алгоритмов поиска подстроки в строке, особенно в случаях, когда подстрока относительно большая по сравнению со строкой, и когда подстрока не совпадает с большинством символов в строке. В лучшем случае, алгоритм Бойера-Мура может работать за время O(n/m), где n - длина строки, а m - длина подстроки.
Адаптивность: Алгоритм Бойера-Мура может адаптироваться к различным языкам и символьным системам, и может использовать различные эвристики для улучшения своей эффективности. Это делает алгоритм Бойера-Мура более гибким и универсальным, чем другие алгоритмы поиска подстроки.
Возможность поиска нескольких вхождений: Алгоритм Бойера-Мура позволяет находить все вхождения подстроки в строке, а не только первое. Это очень полезно в некоторых приложениях, например, при поиске всех вхождений слова в документе.

Недостатки:

Неустойчивость к случайному расположению символов: Алгоритм Бойера-Мура работает лучше, когда подстрока не совпадает с большинством символов в строке. Если же подстрока часто повторяется в строке, то алгоритм Бойера-Мура может работать медленнее, чем другие алгоритмы.
Дополнительная память: Алгоритм Бойера-Мура требует дополнительной памяти для хранения таблицы смещений и таблицы суффиксов. Это может быть проблемой в приложениях с ограниченной памятью.
Сложность реализации: Реализация алгоритма Бойера-Мура может быть сложной и требует тщательного понимания алгоритма и его эвристик. Это может быть проблемой для начинающих разработчиков или для тех, кто не имеет достаточного опыта в программировании.

#### 6-7. Реализовать алгориты согласно номеру индивидуального варианта. Протестировать корректность реализации алгоритма

```python
def boyer_moore(text, pattern):
    # Вычисление таблицы смещений
    table = {}
    n = len(pattern)
    for i in range(n - 1):
        table[pattern[i]] = n - i - 1

    # Поиск подстроки
    m = len(text)
    i = n - 1
    while i < m:
        j = n - 1
        while text[i] == pattern[j]:
            if j == 0:
                return i
            i -= 1
            j -= 1
        i += max(table.get(text[i], n), n - j)
    return -1

def test_boyer_moore():
    assert boyer_moore('hello world', 'world') == 6
    assert boyer_moore('abacadabrabracabracadabrabrabracad', 'rab') == 8
    assert boyer_moore('abacadabrabracabracadabrabrabracad', 'abracadabra') == 14
    assert boyer_moore('abacadabrabracabracadabrabrabracad', 'foo') == -1
```

```python
import re

def boyer_moore_search(text, pattern):
    n = len(text)
    m = len(pattern)
    if m == 0:
        return 0

    skip = [m] * 256

    for i in range(m - 1):
        skip[ord(pattern[i])] = m - i - 1

    i = m - 1

    while i < n:
        j = m - 1
        while text[i] == pattern[j]:
            if j == 0:
                return i
            i -= 1
            j -= 1
        i += max(skip[ord(text[i])], m - j)

    return -1

def count_word_occurrences(file_path, word):
    with open(file_path, 'r') as file:
        text = file.read()

    pattern = r'\b{}\b'.format(re.escape(word))
    matches = re.findall(pattern, text)

    return len(matches)

file_path = '/Users/user/VS.txt'
word = 'Любовь Андреевна'

count = count_word_occurrences(file_path, word)

print('The word "{}" occurs {} times in the file.'.format(word, count))
```

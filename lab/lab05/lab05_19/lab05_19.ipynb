## Задание 2.

Дан лабиринт. Вход в лабиринт находится сверху слева, выход - снизу справа. 
Реализовав алгоритм в соответствии с заданием построить оптимальный путь через лабиринт.

| Вариант | Лабиринт | Алгоритм |
|:---|:---|:---|
| 19  | ![](./img/maze10/maze10_19.png) | Двухлучевой алгоритм |

Манхэттенское расстояние:  $d = |x_2 - x_1| + |y_2 - y_1| $ (Двигаться можно только по горизонтали и вертикали)
​
​

```python
import copy

from PIL import Image

def print_way(maze, A_beam):
    print("__________________________________________")
    for i in range(len(maze)):
        for j in range(len(maze[i])):
            if tuple([i, j]) in A_beam:
                print(u"\u2705", end="")
            elif maze[i][j]:
                print(u"\u2B1C", end="")
            else:
                print(u"\u2B1B", end="")
        print()


# Открываем изображение
img = Image.open("img/maze10/maze10_19.png")
image = img.resize((21, 21))
# image.show()
# Преобразуем в черно-белое
bw_image = image.convert("L")

# Получаем размер изображения
width, height = bw_image.size

# Создаем матрицу размером с изображение, заполненную False
matrix = [[False for _ in range(width)] for _ in range(height)]

# Проходим по каждому пикселю в изображении
for y in range(height):
    for x in range(width):
        # Получаем значение яркости пикселя (от 0 до 255)
        pixel = bw_image.getpixel((x, y))

        # Если яркость больше половины максимальной, то пиксель считаем белым (True)
        if pixel > 127:
            matrix[y][x] = True

print_way(matrix, [])

# двухлучевой поиск

maze = copy.deepcopy(matrix)
A = tuple([1, 0])
B = tuple([19, 20])
# 4 threads mb

def go_beam(maze, point, point_2, x, y):
    A_beam = []
    A_beam.append(point)
    for _ in range(21 + 21):
        x_coord, y_coord = A_beam[-1]
        if tuple([x_coord, y_coord]) == point_2:
            print_way(maze, A_beam)
            return True
        if maze[x_coord][y_coord + y]:
            A_beam.append(tuple([x_coord, y_coord + y]))
        elif maze[x_coord + x][y_coord]:
            A_beam.append(tuple([x_coord + x, y_coord]))
        else:
            print_way(maze, A_beam)
            flag = False
            for _ in range(len(A_beam)):
                bad_x, bad_y = A_beam.pop()
                maze[bad_x][bad_y] = False
                for _ in range(21 + 21):
                    if not A_beam:
                        break
                    x_coord, y_coord = A_beam[-1]
                    if maze[x_coord + x][y_coord]:
                        A_beam.append(tuple([x_coord + x, y_coord]))
                        flag = True
                    elif maze[x_coord][y_coord + y]:
                        A_beam.append(tuple([x_coord, y_coord + y]))
                        flag = True
                    else:
                        if flag:
                            print_way(maze, A_beam)
                            flag = False
                        break
            break
    return False


go_beam(maze, A, B, 1, 1)
go_beam(maze, B, A, -1, -1)
```

Работу алгоритма начнем от начального элемента. При этом вокруг начального элемента рассмотрим 8-ми элементную окрестность. От каждого элемента окрестности до конечного элемента оценим длинну пути. При этом расстояние между точками вычислется по формуле: $d = |x_2 - x_1| + |y_2 - y_1| $
Таким образом вычисляем восемь значений из которых выбираем минимальное. Элемент от которого расстояние оказалось минимальным выбирается в качестве элемента трассы. Вокруг него снова рассматривается 8-ми элементная окрестность. Продолжаем процесс до тех пор пока не будет достигнут конечный элемент.
1. |8-2| + |2-9| = 13 и |7-2| + |2-9| = 12 новая трасса имеет координаты (7, 2)
2. |6-2| + |2-9| = 11 и |6-2| = |3-9| = 10 и |8-2| + |3-9| = 12 новая трасса имеет координаты (6, 3)
3. |6-2| + |4-9| = 9 новая трасса имеет координаты (6, 4)
4. У этой трассы нет пути, возвращаемся и выбираем другую трассу
5. Новая трасса имеет координаты (8,3)
6. |8-2| + |4-9| = 11 новая трасса имеет координаты (8, 4)
7. |8-2| + |5-9| = 10 новая трасса имеет координаты (8, 5)
8. |8-2| + |6-9| = 9 новая трасса имеет координаты (8, 6)
9. |8-2| + |7-9| = 8 новая трасса имеет координаты (8, 7)
10. |8-2| + |8-9| = 7 и |7-2| + |8-9| = 6 новая трасса имеет координаты (7, 8)
11. |6-2| + |8-9| = 5 и |6-2| + |7-9| = 6 новая трасса имеет координаты (6, 8)

```python
from PIL import Image

# Открываем изображение
img = Image.open("./img/maze10_09.png")
image = img.resize((21, 21))
#image.show()
bw_image = image.convert("L")
width, height = bw_image.size
matrix = [[False for _ in range(width)] for _ in range(height)]
for y in range(height):
    for x in range(width):
        pixel = bw_image.getpixel((x, y))
        if pixel > 127:
            matrix[y][x] = True

for i in matrix:
    print('[', end="")
    for j in i:
        if j:
            print("0 ", end="")
        else:
            print("1 ", end="")
    print(']', end="")
    print()

```

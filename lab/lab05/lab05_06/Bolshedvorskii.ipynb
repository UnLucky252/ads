```python
from copy import copy

o = -1
_ = None
maze = [
    [o, o, o, o, o, o, o, o, o, o, o, o, o, o, o, o, o, o, o, o, o],
    [_, _, _, _, _, _, _, _, _, _, _, _, o, _, _, _, o, _, _, _, o],
    [o, _, o, o, o, o, o, o, o, o, o, _, o, _, o, _, o, _, o, o, o],
    [o, _, _, _, _, _, o, _, o, _, _, _, o, _, o, _, o, _, _, _, o],
    [o, o, o, o, o, _, o, _, o, o, o, o, o, _, o, _, o, o, o, _, o],
    [o, _, _, _, o, _, o, _, _, _, _, _, _, _, o, _, _, _, _, _, o],
    [o, o, o, _, o, _, o, _, o, _, o, o, o, o, o, o, o, o, o, _, o],
    [o, _, _, _, o, _, o, _, o, _, _, _, _, _, _, _, o, _, o, _, o],
    [o, _, o, _, o, _, o, _, o, o, o, o, o, o, o, _, o, _, o, _, o],
    [o, _, o, _, o, _, o, _, _, _, _, _, _, _, o, _, o, _, _, _, o],
    [o, _, o, o, o, _, o, o, o, _, o, o, o, o, o, _, o, o, o, o, o],
    [o, _, _, _, o, _, o, _, _, _, o, _, _, _, o, _, _, _, _, _, o],
    [o, o, o, _, o, _, o, o, o, _, o, _, o, _, o, o, o, o, o, _, o],
    [o, _, _, _, o, _, _, _, o, _, o, _, o, _, _, _, o, _, _, _, o],
    [o, _, o, o, o, o, o, _, o, _, o, _, o, o, o, _, o, _, o, o, o],
    [o, _, o, _, _, _, _, _, o, _, o, _, o, _, o, _, o, _, _, _, o],
    [o, _, o, _, o, o, o, o, o, o, o, _, o, _, o, _, o, _, o, _, o],
    [o, _, _, _, o, _, _, _, _, _, o, _, o, _, _, _, o, _, o, _, o],
    [o, _, o, o, o, o, o, _, o, _, o, _, o, _, o, o, o, o, o, _, o],
    [o, _, _, _, _, _, _, _, o, _, _, _, o, _, _, _, _, _, _, _, _],
    [o, o, o, o, o, o, o, o, o, o, o, o, o, o, o, o, o, o, o, o, o],
]


def escape_trace(maze: list[list], beg: tuple, end:tuple)->list:
    if len(beg) != 2 or len(end) != 2:
        raise ValueError('beg and end must be like (x, y)')
    
    x_beg, y_beg = beg
    x_end, y_end = end
    
    # Wave front spreading
    maze[y_beg][x_beg] = 0
    front = [beg]
    front_grade = 0
    while end not in front:
        front_grade += 1
        spread = []
        for cell in front:
            x, y = cell
            spread_pattern = [
                (x + 1, y), 
                (x - 1, y), 
                (x, y + 1), 
                (x, y - 1)
            ]
            for thing in spread_pattern:
                x_spr, y_spr = thing
                if maze[y_spr][x_spr] is None:
                    spread.append(thing)
                    maze[y_spr][x_spr] = front_grade
        front = copy(spread)
        
    # Traceback
    trace = [end]
    while end != beg:
        front_grade -= 1
        x, y = end
        spread_pattern = [
            (x + 1, y), 
            (x - 1, y), 
            (x, y + 1), 
            (x, y - 1)
        ]
        for thing in spread_pattern:
            x_spr, y_spr = thing
            if x_spr > 20 or y_spr > 20:
                continue
            if maze[y_spr][x_spr] == front_grade:
                end = thing
                trace.append(end)
                break
    return trace


# Testing
trace = escape_trace(maze, (0, 1), (20, 19))

maze_size = len(maze)
for y in range(maze_size):
    row = []
    for x in range(maze_size):
        el = ' '
        if (x, y) in trace:
            el = '.'
        elif maze[y][x] is None:
            el = ' '
        elif maze[y][x] == -1:
            el = '#'
        row.append(el)
    print(' '.join(row))
    
```
